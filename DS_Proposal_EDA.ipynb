```python
pruned_data = data.drop(columns = ['OBJECTID', 'FPA_ID', 'SOURCE_SYSTEM',
       'NWCG_REPORTING_UNIT_ID',
       'NWCG_REPORTING_UNIT_NAME', 'SOURCE_REPORTING_UNIT',
       'SOURCE_REPORTING_UNIT_NAME', 'LOCAL_FIRE_REPORT_ID',
       'LOCAL_INCIDENT_ID', 'FIRE_CODE', 'FIRE_NAME',
       'ICS_209_INCIDENT_NUMBER', 'ICS_209_NAME', 'MTBS_ID', 'MTBS_FIRE_NAME',
       'COMPLEX_NAME', 'STAT_CAUSE_CODE','OWNER_CODE', 'FIPS_NAME','FIPS_CODE', 'Shape'])
```

```python
pruned_data.head()
```

```python
columns = ['SOURCE_SYSTEM_TYPE', 'NWCG_REPORTING_AGENCY', 'STAT_CAUSE_DESCR','FIRE_SIZE_CLASS','OWNER_DESCR', 'STATE']

#NWCG_REPORTING_AGENCY is not satisfying both ANOVA assumptions
```

```python
#checking which featuers are satisfying anova assumptions - normality check
for column in columns:
  sub_classes = pruned_data[column].unique()
  for classs in sub_classes:
    data = pruned_data[pruned_data[column] == classs]
    if len(data) > 500:
      data = data.sample(n=500,random_state=2)
      data['DISCOVERY_TIME'].hist()
      plt.title(classs)
      plt.show()
```

```python
# STD deviation anova assumption check
for column in columns:
  #sub_classes = pruned_data[column].unique()
  #data = pruned_data[pruned_data[column] == classs]
  #data['DISCOVERY_TIME'].std()
  #data = pruned_data.sample(n=1000,random_state = 32)
  stdn = pruned_data.groupby(column)['DISCOVERY_TIME'].std()
  print(column)
  print(stdn.max()/stdn.min())
```

All of the above columns are satisfying the std assumtion

```python
column = 'NWCG_REPORTING_AGENCY'
sub_classes = pruned_data[column].unique()
print(sub_classes)
```

```python
#perofrming the anova test to find the p value
temp_data = pruned_data.dropna(subset=['DISCOVERY_TIME'],inplace=False)
columns = ['STAT_CAUSE_DESCR','FIRE_SIZE_CLASS','OWNER_DESCR','STATE','SOURCE_SYSTEM_TYPE']
for column in columns:
  x = []
  for classs in temp_data[column].unique():
    x.append(list(temp_data[temp_data[column] == classs].DISCOVERY_TIME))
  fvalue, pvalue = stats.f_oneway(*x)
  print(column,fvalue, pvalue)
#all columms tested above have statistically significant influence on Discovery time
# We will be using these five columns to impute Discovery Time
```

```python
# dropping na data for now to perform EDA
for column in pruned_data.columns:
  if (pruned_data[column].isna().any()):
    print(column,'is na ',pruned_data[column].isna().sum()/len(pruned_data),'')
# count of na datapoints
for column in pruned_data.columns:
  if (pruned_data[column].isna().any()):
    print(column,'is na ',pruned_data[column].isna().sum(),'')
```

```python
import calendar

def JulianDate_to_MMDDYY(row):
    month = 1
    day = 0
    while row['DISCOVERY_DOY'] - calendar.monthrange(row['FIRE_YEAR'],month)[1] > 0 and month <= 12:
        row['DISCOVERY_DOY'] = row['DISCOVERY_DOY'] - calendar.monthrange(row['FIRE_YEAR'],month)[1]
        month = month + 1
    return month
```

```python
temp_data['MONTH'] = temp_data.apply(JulianDate_to_MMDDYY, axis = 1)
```
